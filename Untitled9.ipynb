```python
columns_of_interest = ['Age', 'Fare', 'SibSp']

# Load the Titanic dataset
data = pd.read_csv('/content/titanic_dataset.csv')

# Step 1: Visualize the distribution of each column to identify outliers
for column in columns_of_interest:
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

# Step 2: Identify outliers using a threshold (e.g., 3 standard deviations from the mean)
for column in columns_of_interest:
    mean = data[column].mean()
    std = data[column].std()
    threshold = 3 * std
    outliers = data[(data[column] < mean - threshold) | (data[column] > mean + threshold)]
    print(f'Number of outliers in {column}: {len(outliers)}')

# Step 3: Handle outliers (e.g., replace with the median value)
for column in columns_of_interest:
    median = data[column].median()
    mean = data[column].mean()
    std = data[column].std()
    threshold = 3 * std
    data[column] = np.where((data[column] < mean - threshold) | (data[column] > mean + threshold), median, data[column])

# Step 4: Visualize the distribution after handling outliers
for column in columns_of_interest:
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column} (After Handling Outliers)')
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming 'data' contains your entire dataset

# Selecting only numeric columns for scaling
numeric_columns = data.select_dtypes(include=[np.number]).columns

# Separate the target variable ('Survived') from the feature set
X = data[numeric_columns].drop('Survived', axis=1)
y = data['Survived']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the feature set
X_scaled = scaler.fit_transform(X)

# Convert the scaled feature set back to a DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Concatenate the scaled feature set with the target variable
scaled_data = pd.concat([X_scaled_df, y], axis=1)

# Check the scaled dataset
print(scaled_data.head())
```
